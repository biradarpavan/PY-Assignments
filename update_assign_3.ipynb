```python
class Product:
    products = []

    def __init__(self, name, category, price, quantity, expiry_date):
        self.name = name
        self.category = category
        self.price = price
        self.quantity = quantity
        self.expiry_date = self.exp_date(expiry_date)

    def exp_date(self, date_str):
        date = date_str.split('-')
        day = int(date[0])
        month = int(date[1])
        year = int(date[2])
        return (day, month, year)

    def expiry(self, today):
        day, month, year = self.expiry_date
        expiry_datetime = datetime(year, month, day)
        return expiry_datetime < today


def add_product():
    name = input("Enter product name: ")
    category = input("Enter category: ")
    price = float(input("Enter price: "))
    quantity = int(input("Enter quantity: "))
    expiry_date = input("Enter expiry date DD-MM-YYYY: ")
    p = Product(name, category, price, quantity, expiry_date)
    Product.products.append(p)
    print(f"Product {name}, {category} added to inventory.")


def remove_product(p_name):
    removed = False
    for product in Product.products[:]:
        if product.name == p_name:
            Product.products.remove(product)
            removed = True
    if removed:
        print(f"Product'{p_name}'removed.")
    else:
        print(f"Product'{p_name}'not found.")


def search(s):
    match = False
    for product in Product.products:
        if s.lower() in product.name.lower() or s.lower() in product.category.lower():
            print(disp(product))
            match = True
    if not match:
        print("No product found")


def list():
    if not Product.products:
        print("No product found")
    else:
        for product in Product.products:
            print(disp(product))


def category():
    dict = {}
    for i in Product.products:
        key = i.category
        if key in dict:
            dict[key].append(i)
        else:
            dict[key] = [i]
    for category, products in dict.items():
        print(f"Category: {category}")
        for i in products:
            print(f"{disp(i)}")


def remove_exp():
    today = datetime.now()
    expired = [i for i in Product.products if i.expiry(today)]
    for i in expired:
        Product.products.remove(i)
    print("Expired products removed")


def save_to_file(filename):
    try:
        file = open(filename,'a')
        for p in Product.products:
            file.write(f"{p.name},{p.category},{p.price},{p.quantity},{p.expiry_date[0]:02}-{p.expiry_date[1]:02}-{p.expiry_date[2]}\n")
        print("Saved to file.")
    except IOError as e:
        print(f"Error: {e}")


def load_from_file(filename):
    try:
        file = open(filename,'r')
        for line in file:
            name, category, price, quantity, expiry_date = line.strip().split(',')
            Product.products.append(Product(name, category, float(price), int(quantity), expiry_date))
        print("Inventory loaded from file.")
    except IOError as e:
        print(f"Error: {e}")


def disp(p):
    return f"{p.name}, {p.category}, {p.price}, {p.quantity}, {p.expiry_date[0]:02}-{p.expiry_date[1]:02}-{p.expiry_date[2]}"


def main():
    load_from_file('inventory.txt')

    while True:
        print("\nInventory Management System")
        print("1. Add a product")
        print("2. Remove a product")
        print("3. Search for products")
        print("4. List all products")
        print("5. Categorize products")
        print("6. Check for expired products")
        print("7. Save inventory to file")
        print("8. Exit")

        choice = input("Enter choice: ")

        if choice == '1':
            add_product()
        elif choice == '2':
            p_name = input("Enter product name to remove: ")
            remove_product(p_name)
        elif choice == '3':
            term = input("Enter search term (name or category): ")
            search(term)
        elif choice == '4':
            list()
        elif choice == '5':
            category()
        elif choice == '6':
            remove_exp()
        elif choice == '7':
            save_to_file('inventory.txt')
        elif choice == '8':
            print('Exiting... from inventory!!!')
            save_to_file('inventory.txt')
            break
        else:
            print('Invalid Choice!!!')

main()
```
